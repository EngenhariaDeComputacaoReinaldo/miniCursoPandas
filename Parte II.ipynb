## **Agrupamento**
GroupBy essencialmente divide o dado dentro de diferentes grupos, dependendo da variável de sua escolha. Por exemplo a expressão dfCall.groupby('month') ira dividir nosso DataFrame atual em meses.

A função groupby() retorna um objeto groupby, mas, essencialmete descreve como as linhas de um dataSet original podem ser divididas.

Seus Grupos essencialmente se tornam índices de subgrupos de dados

### **retornar os indices dos agrupamentos**

```
dfCall.groupby(['month']).groups.keys()
```

### **retorna  a primeira entrada para cada mês**

```
dfCall.groupby('month').first()
```

### **Soma do tempo de ligações por mês**

```
dfCall.groupby('month')['duration'].sum()
```

### **Quantidade de Datas por Mês**

```
dfCall.groupby('month')['date'].count()
```

### **Qual a quantidade de chamadas, textos e dados são enviados por mes, dividos por rede?**

```
dfCall.groupby(['month', 'network_type'])['date'].count()
```

Agregações permitem que várias estátisticas sejam calculdas por um grupo em um calculo apenas. 
Instruções de agregaçaõ são fornecidas na forma de dicionarios ou listas em Python. No caso de dicionários, as chaves são usadas para especificar as colunas na qual a operação será realizada e os valores especificam a função a ser aplicada.

```
dfCall.groupby(['month', 'item']).agg(
    {
         'duration':"sum",    # Soma duração da ligação por grupo
         'network_type': "count",  # Obtém a contagem de redes (quantidade de registros)
         'date': "first"  # Obtem a primeira data por grupo
    }
)
```

### **Definindo a função de agregação antes**

```
aggCustom = {
    # find the min, max, and sum of the duration column
    'duration': [min, max, sum],
     # find the number of network type entries
    'network_type': "count",
    # min, first, and number of unique dates per group
    'date': [min, 'first', 'nunique']
}
```

### **Aplicando sobre  DataSet**

```
# Group the data frame by month and item and extract a number of stats from each group
dfCall.groupby(
    ['month', 'item']
).agg(aggCustom)
```
